```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.0,10.0,0.1)
y = 2*x+3

y_noise = 2* np.random.normal(size = x.size)
y_data = y+y_noise

plt.plot(x,y_data,'ob')
plt.plot(x,y,'r')
```

```python
x = np.arange(-5.0,5.0,0.1)
y = x**3+x**2+x+3
y_noise = 20*np.random.normal(size = x.size)
y_data = y+y_noise
plt.plot(x,y_data,'ob')
plt.plot(x,y,'r')
```
