# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,5,3,9,4,12],[2,3,4,5,7,8])
plt.ylabel("funny number")
```

```python
plt.plot([1,2,3,4],[4,5,7,-2],'y*')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2,'r--',t,t,'ys-',t,t**3,'g+-')
```
