```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("china_gdp.csv")
```

```python
xx = max(df["Year"])
yy = max(df["Value"])
x_data, y_data = np.asanyarray(df["Year"]/max(df["Year"])), np.asanyarray(df["Value"]/max(df["Value"]))
```

```python
def sigmoid (x, beta1, beta2):
    y = 1/(1+np.exp(-beta1*(x-beta2)))
    return y
```

```python
from scipy.optimize import curve_fit

popt, pcov = curve_fit(sigmoid,x_data,y_data)
```

```python
x = x_data
y = sigmoid(x,*popt)
plt.plot(x,y,'r')
plt.plot(x_data,y_data,'ob')
```
