```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=150, centers=[[2, 4], [6, 9], [4, 5]], cluster_std=0.5)
plt.scatter(x[:,0], x[:,1], c = y)
```

```python
model = DBSCAN(eps=0.4, min_samples=7)
model.fit(x)
```

```python
plt.scatter(x[:,0], x[:,1], c = model.labels_)
```
